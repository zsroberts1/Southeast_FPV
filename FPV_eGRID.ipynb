This code takes EGRID 2023 plant data, isolates Southeast solar plants, and exports the sum of capacity and generation by state to an Excel sheet.

It also calculates other useful data like total generation and capacity (overall and for solar) and percent of overall generation and capacity provided by solar.

Prepping EGRID 2023 plant data.

```python
df_plant23=pd.read_excel(io="/content/gdrive/MyDrive/DATA/egrid2023_data_rev1.xlsx",sheet_name="PLNT23",skiprows=1,keep_default_na=False)
```

```python
pd.set_option('display.max_columns',None)
```

```python
df_plant23.head()
```

```python
capacities=df_plant23["NAMEPCAP"]
```

```python
df_plant23['PLPRMFL'].value_counts() # Count of plants for each generation type
```

```python
# Total US solar capacity
Total_Capacity_Sun=df_plant23.loc[df_plant23['PLPRMFL'] == 'SUN', 'NAMEPCAP'].sum()
print("The total solar plant installed capacity in 2023 was",Total_Capacity_Sun,"MW")
```

```python
# Total Southeast solar capacity
Total_Capacity_SE=df_plant23.loc[(df_plant23['PLPRMFL'] == 'SUN') & (df_plant23['PSTATABB'].isin(['NC','SC','GA','TN','KY','LA','AL','AR','MS','FL'])), 'NAMEPCAP'].sum()
```

```python
# Percentage of solar capacity located in Southeast
SE_Capacity_Pct= Total_Capacity_SE/Total_Capacity_Sun
print("Percentage of US solar capacity located in the Southeast is",SE_Capacity_Pct*100,"%")
```

```python
df_plant23.PLNGENAN=pd.to_numeric(df_plant23.PLNGENAN)
```

```python
# Total US generation in MW
Total_Generation=df_plant23.PLNGENAN.sum()
print("Total US power generation in 2023 is",Total_Generation,"MW")
```

```python
# Total US solar generation in MW
Total_Generation_Sun=df_plant23.loc[df_plant23['PLPRMFL'] == 'SUN', 'PLNGENAN'].sum()
```

```python
# Total Southeast solar generation in MW
Total_Generation_Sun_SE=df_plant23.loc[(df_plant23['PLPRMFL'] == 'SUN') & (df_plant23['PSTATABB'].isin(['NC','SC','GA','TN','KY','LA','AL','AR','MS','FL'])), 'PLNGENAN'].sum()
```

```python
# Percentage of solar generation located in Southeast
SE_Solar_Pct= Total_Generation_Sun_SE/Total_Generation_Sun
print("Percentage of US solar generation located in the Southeast is", SE_Solar_Pct*100,"%")
```

```python
# Subset of all solar plants
df_solar23=df_plant23[(df_plant23.PLFUELCT=="SOLAR")]
```

```python
df_solar23
```

```python
NumSolarPlants=len(df_solar23)
```

```python
NumSolarPlants
```

```python
# Subset of Southeast solar plants only
df_solar23_SE=df_solar23[df_plant23['PSTATABB'].isin(['NC','SC','GA','TN','KY','LA','AL','AR','MS','FL'])]
```

```python
df_solar23_SE
```

```python
# Pivot table for sum of solar capacity and generation by state in 2023 in MW
Capacity_SE=pd.pivot_table(df_solar23_SE,values=['NAMEPCAP','PLNGENAN'],index='PSTATABB',aggfunc='sum',fill_value="")
Capacity_SE
```

```python
Capacity_SE.to_excel("/content/gdrive/MyDrive/DATA/SE_terrestrial_solar.xlsx")
```
